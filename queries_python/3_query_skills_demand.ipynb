Analizar y graficar las skills mas demandadas para trabajos de data analista como trabajo remoto

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
from datasets import load_dataset

#loading data
dataset = pd.read_csv("3_query_skills_demand.csv", sep=',')
```

```python
dataset
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=dataset, x='demand_count', y='skill_name',  palette='dark:b_r', legend=False)
sns.despine()
plt.title('Demanda de Skills Remoto para Data Analyst Jobs')
plt.xlabel('Conteo de Skills')
plt.ylabel('')
plt.show()
```
