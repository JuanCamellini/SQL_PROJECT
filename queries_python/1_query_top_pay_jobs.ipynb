Graficar los trabajos mejores remunerados de data analista

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
from datasets import load_dataset

#loading data
dataset = pd.read_csv("csv/1_query_top_pay_jobs.csv", sep=',')
```

```python
dataset
df_pay_jobs = dataset[['job_title_short', 'salary_month','name']]
df_pay_jobs.reset_index(drop=True)
```

```python
sns.set_theme(style='ticks')

sns.barplot(
    data=df_pay_jobs,
    x='name',
    y='salary_month',
    hue='name',
    palette='dark:b_r',
    legend=False
)
ay = plt.gca()

ay.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))



sns.despine()
plt.tight_layout()
plt.title('Top 10 Trabajos Remunerados Por Mes de Data Analista')
plt.ylabel('')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.figure(figsize=(17,10))
```
