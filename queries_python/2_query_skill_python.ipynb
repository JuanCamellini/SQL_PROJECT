Analizar el conteo de las skills del csv obtenido de la 2 query de los trabajos mejores pagos de data analista remotos

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
from datasets import load_dataset

#loading data
dataset = pd.read_csv("2_query_top_10_jobs_payed_skills.csv", sep=',')
```

```python
# Elegimos las 7 skills que mas se repiten 
da_skills = dataset['skills'].value_counts().head(7).to_frame()
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=da_skills, x='count', y='skills', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Conteo de Skills en los Top-50-Pay Trabajos')
plt.xlabel('Conteo de Skills')
plt.ylabel('')
plt.show()
```
